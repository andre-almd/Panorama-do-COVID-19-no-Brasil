# Panorama do COVID-19 no Brasil

COVID-19 é uma doença infecciosa causada por coronavírus, descoberto no final de 2019.

Transmitido principalmente por meio de gotículas provenientes de tosses ou espirros de pessoas infectadas, a gravidade dos sintomas varia muito de pessoa para pessoa.

Fato é, não se sabe muita coisa a respeito do COVID-19. Estudos estão sendo realizados no mundo todo, porém os resultados ainda não são conclusivos e definitivos.

<center><img alt="Colaboratory logo" width="50%" src="https://raw.githubusercontent.com/carlosfab/dsnp2/master/img/covid19.jpg"></center>

Até o presente momento, observa-se que cerca de 80% dos casos confirmados são assintomáticos e rápidos. A maioria das pessoas que se encaixam nesse grupo, se recupera sem nenhuma sequela.

No entanto, 15% das pessoas terão infecções graves e precisarão de oxigênio. O restante das pessoas, que representam 5%, serão classificadas como infecções muito graves e precisarão de ventilação assistida, por meio de respiradores mecânicos em ambiente hospitalar.

Com o objetivo de elevar a consciência situacional a respeito do COVID-19 no Brasil, irei realizar uma análise sobre os dados públicos da doença.

## Obtenção dos Dados

Os dados utilizados neste projeto são mantidos pelo **Our World in Data (OWID)**. 

O [OWID](https://ourworldindata.org/) é uma publicação científica online que se concentra em grandes problemas globais, como pobreza, doenças, fome, mudanças climáticas, guerra, riscos existenciais e desigualdade. Seu objetivo é tornar o conhecimento sobre os grandes problemas acessível e compreensível.

**Como descrito bem no site do projeto, a missão do OWID é publicar a “pesquisa e dados para avançar contra os maiores problemas do mundo”.**

Os dados esão disponíveis [aqui](https://github.com/owid/covid-19-data/tree/master/public/data) e são atualizados diariamente, enquanto durar a pandemia.

A fonte principal dos dados utilizados pelo OWID é o **Centro de Ciência e Engenharia de Sistemas (CSSE) da Universidade Johns Hopkins**. O conjunto de dados dos casos registrados e óbitos é atualizado diariamente.

---
**Notas importantes:**

1 - Os dados referentes a esta aálise foram baixados no dia **16/08/2022**.

2 - Os dados de casos confirmados e óbitos são coletados pela Johns Hopkins University por data de relatório, em vez de data de teste/óbito. Portanto, o número que eles relatam em um determinado dia não representa necessariamente o número real naquela data, devido à longa cadeia de notificação que existe entre um novo caso/óbito e sua inclusão nas estatísticas. Isso também significa que as séries temporais podem mostrar mudanças repentinas (negativas ou positivas) quando um país corrige dados históricos, porque anteriormente havia sub ou superestimado o número de casos/óbitos.

**Como saber se esses dados são relevantes???**

Vamos gerar uma regressão com os principais atributos que possuem correlação com a quantidade de mortes e analisar o nosso teste de hipótese da significância dos dados.

Aqui, a hipótese nula diz que os atributos são = 0, ou seja, não representativos para determinar o resultado da nossa variável dependendte, quantidade de mortes. Se a hipótese for rejeitada, temos que os atributos são sifnificativos e não podemos desprezá-los.


```
import statsmodels.api as sm

y = df_new['total_deaths_per_million']

X = df_new[['female_smokers', 'median_age','human_development_index', 'stringency_index', 'life_expectancy', 'gdp_per_capita', 'cardiovasc_death_rate']]

X = sm.add_constant(X)

est = sm.OLS(y, X, missing='drop').fit()

print(est.summary())
```

Com os resultados acima pode ser visto que, **a nível de significância de 5%**, **os principais atributos não podem ser considerados nulos** e as correlações são relevantes na estimativa de mortes por milhão de pessoas.

É importante lembrar que a maioria das correlações não são fortes, isso explica o r-quadrado baixo da análise.

Nos testes foram testadas os atributos dos quais as correlações foram mais interessantes: 
- female_smokers: percentual de mulheres fumantes
- median_age: idade média da população
- human_development_index: IDH da população
- stringency_index: índice de resposta do governo
- life_expectancy: expectativa de vida
- gdp_per_capita: PIB do país
- cardiovasc_death_rate: taxa de morte por problemas cardiovasculares

## Conclusão

Analisra os dados da Covid-19 não é fácil, pois os dados são muito muitos e há muitas informações faltantes como foi visto. Isso pode depender do país, da política governamental e não entramos nesses detalhes para entender como cada país divulga os seus dados.

No dataset analisado foi percebido que há uma relação direta entre alguns dados populacionais e a quantidade de mortes, e essa relação não pode ser desprezada:

- Foi verificado que países com uma idade média alta tende a ter mais mortes;
- Países com um alto IDH e PIB tenderam a ter mais mortes. Aqui é um fator interesante, talvez o motivo pode ser explicado pelo fato desses países terem grandes populações, ter um alto fluxo de pessoas entrando e saindo do país, estarem em regiões mais propoensas a transmissão do vírus, entre outro fatores;
- Países que tomaram medidas mais drásticas de segurança não evitaram o problema, mas conseguiram ter uma redução na quantidade de mortes por milhão de pesoas.

Com a análise de hipótese para os atributos populacionais foi verificado que eles são estatísticamente importantes em uma modelagem da quantidade de mortes por milhão de pessoas e não podem ser desprezados.

No Brasil, a quantidade de casos e mortes seguiu uma uma padão parecido com o padão mundial, porém com uma curva mais intensa nos anos de 2020 e 2021. Talvez pelas ações tomadas pelo país durante esse período...

É visível que a partir do momento em que as vacinas começaram a ser aplicadas a quantidade de novos casos e mortes começa a reduzir significativamente.

Em 2022 há um novo pico de casos no início do ano, percebido nos dados mundiais também. Um fator que pode estar relacionado a isso são as festas de final de ano e, aqui no Brasil, a época de carnaval no início do ano. Porém a quantidade de mortes não é tão afetada, talvez pelo fator das pessoas já estarem vacinadas.
